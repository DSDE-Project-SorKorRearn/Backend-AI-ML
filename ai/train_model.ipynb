```python
import pandas as pd
```

```python
df = pd.read_csv("../data/processed_traffy_data_for_visualization.csv")
```

```python
df
```

```python
df[['lon','lat']] = df['coords'].str.split(',', expand=True)
```

```python
df.info()
```

```python
X = df[["lat", "lon"]]
```

```python
from sklearn.cluster import KMeans
```

```python
kmeans = KMeans(n_clusters=3, random_state=42, algorithm='elkan')

df["cluster_label"] = kmeans.fit_predict(X)
```

```python
df[["lat","lon","cluster_label"]]
```

```python
import ast
df['type'] = df['type'].apply(ast.literal_eval)
```

```python
df = df.explode("type", ignore_index=True)
```

```python
df
```

```python
import pandasql

cluster_df = pandasql.sqldf("""
            with query_cluster as (
            select cluster_label, type, count(*) as count_type, row_number() over (partition by cluster_label order by count(*) desc) as rn
            from df group by cluster_label, type order by cluster_label, count_type desc
            ),
            district_count_with as (
                select cluster_label,type,district,count(*) as district_count, row_number() over (partition by cluster_label,type order by count(*) desc) as rn from df group by cluster_label,type,district
            ),
            district_top_1 as (
            select * from district_count_with where rn = 1
            )
            select q.cluster_label,q.type as top3_type, q.count_type as top3_type_count, d.district as top3_district, d.district_count as top3_district_count  from query_cluster q join district_top_1 d on q.cluster_label = d.cluster_label and q.type = d.type where q.rn <= 3;
            """)
```

```python
cluster_df = cluster_df.groupby(['cluster_label'])[['top3_type', 'top3_type_count','top3_district', 'top3_district_count']].agg(tuple).reset_index()
```

```python
df = df.merge(cluster_df, on="cluster_label", how="left")
```

```python
df
```

```python
clustered_df = df[["coords", "cluster_label", "top3_type", "top3_type_count", "top3_district", "top3_district_count"]].drop_duplicates(inplace=False)
```

```python
clustered_df.to_csv("../data/clustered_traffy.csv", index=False)
```
