```python
import pandas as pd
```

```python
df = pd.read_csv("../data/clean_traffy.csv")
```

```python
df
```

```python
df[['lon','lat']] = df['coords'].str.split(',', expand=True)
```

```python
df.info()
```

```python
district_list = df["district"].unique().tolist()

district_list
```

```python
from sklearn.cluster import KMeans
for district in district_list:
    mask = df["district"] == district
    X = df.loc[mask, ["lat","lon"]]
    kmeans = KMeans(n_clusters=min(3, X.shape[0]), random_state=42)
    df.loc[mask, "cluster_label"] = kmeans.fit_predict(X)
```

```python
import ast
df['type'] = df['type'].apply(ast.literal_eval)
```

```python
df = df.explode("type", ignore_index=True)
```

```python
df
```

```python
import pandasql

cluster_df = pandasql.sqldf("""
            with query_cluster as (
            select cluster_label,district, type, count(*) as count_type, row_number() over (partition by cluster_label,district order by count(*) desc) as rn
            from df group by cluster_label, district, type order by cluster_label,district, count_type desc
            )
            select q.cluster_label,q.district ,q.type as top_type, q.count_type as top_type_count from query_cluster q where q.rn <= 3;
            """)
cluster_df
```

```python
cluster_df = cluster_df.groupby(['cluster_label', 'district'])[['top_type', 'top_type_count']].agg(tuple).reset_index()
```

```python
df = df.merge(cluster_df, on=["cluster_label", "district"], how="left")
```

```python
df
```

```python
clustered_df = df[["coords", "district","cluster_label", "top_type", "top_type_count"]].drop_duplicates(inplace=False)
```

```python
clustered_df.to_csv("../data/clustered_traffy.csv", index=False)
```
